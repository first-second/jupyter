```python
#General Tree

class Treenode:
    def __init__(self,data):
        self.data = data
        self.children = []
        self.parent = None
        
    def addchild(self,child):
        child.parent = self
        self.children.append(child)
        
    def getlevel(self):
        level = 0
        p = self.parent
        while p:
            level += 1
            p = p.parent
        return level
    
    def printtree(self):
        space = " " * 2 * self.getlevel()
        
        print(space + self.data)
        if self.children:
            for child in self.children:
                child.printtree()

def buildtree():
    root = Treenode('Electronics')
    
    tv = Treenode('television')
    tv.addchild(Treenode('SonyTv'))
    tv.addchild(Treenode('SamsungTv'))
    
    mobile = Treenode('mobile')
    mobile.addchild(Treenode('OnePlus'))
    mobile.addchild(Treenode('Iphone'))
    
    root.addchild(tv)
    root.addchild(mobile)
    
    return root
    
    
if __name__ == "__main__":
    
    root = buildtree()
    root.printtree()
```

```python
# binaryTree
class Treenode:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
        
    def addchild(self,child):
        
        if child == self.data:
            return
        
        if child < self.data:
            if self.left:
                self.left.addchild(child)
            else:
                self.left = Treenode(child)
        else:
            if self.right:
                self.right.addchild(child)
            else:
                self.right = Treenode(child)
        
    
    def printinordertree(self):
        element = []
        
        if self.left:
            element += self.left.printinordertree()
            
        element.append(self.data)
        
        if self.right:
            element += self.right.printinordertree()
        
        return element
    
    def printpreordertree(self):
        
        element = []
        
        element.append(self.data)
        
        if self.left: element += self.left.printpreordertree()
        if self.right: element += self.right.printpreordertree()
        
        return element
        
        

def buildtree(lst):
    root = Treenode(lst[0])
    
    for i in range(1,len(lst)):
        root.addchild(lst[i])
    
    return root
    
    
if __name__ == "__main__":
    
    lst = [4,2,1,8,5,3,6]
    root = buildtree(lst)
    
    print(f'inorder : {root.printinordertree()}')
    print(f'preorder : {root.printpreordertree()}')
```
