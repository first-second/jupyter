```python
import pandas as pd
import matplotlib.pyplot as plt
#df = pd.read_csv(file)
df1=pd.read_excel('gold.xlsx',sheet_name='Sheet1')
df2=pd.read_excel('gold.xlsx',sheet_name='Sheet2')
df3=pd.read_excel('gold.xlsx',sheet_name='Sheet3')
print(df1)
```

```python
df1['year'] = df1['date'].dt.strftime('%Y')
df1['month'] = df1['date'].dt.strftime('%m')
df1['D'] = df1['date'].dt.strftime('%d')
count=0
for i in df1['year']:
    if i=="2021":
        count+=1
print(count)
print(df1)
```

```python
df2['year'] = df2['date'].dt.strftime('%Y')
df2['month'] = df2['date'].dt.strftime('%m')
df2['D'] = df2['date'].dt.strftime('%d')
count=0
for i in df2['year']:
    if i=="2022":
        count+=1
print(count)
print(df2)
```

```python
df3['year'] = df3['date'].dt.strftime('%Y')
df3['month'] = df3['date'].dt.strftime('%m')
df3['D'] = df3['date'].dt.strftime('%d')
count=0
for i in df3['year']:
    if i=="2022":
        count+=1
print(count)
print(df3)
```

```python
c=df1['year'].count()+df2['year'].count()+df3['year'].count()
ds = df1['year'].append(df2['year'], ignore_index = True)
ds = ds.append(df3['year'], ignore_index = True)
co=sorted(list(set(ds)))
print(co)
print(c)
```
